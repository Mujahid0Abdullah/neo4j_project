```python
import json

# Verileri bir DataFrame'e dönüştürün
data = pd.read_csv("PERSON_BATTLErelation - الورقة1 (1).csv")

df = pd.DataFrame(data)

# Her bir kahraman için JSON nesnesi oluşturun
kahramanlar = []
for kahraman in df.groupby('pAdi'):
    kahraman_adi = kahraman[1]['pAdi'].iloc[0]
    muharebeler = kahraman[1]['bAdi'].tolist()
    id= kahraman[1]['pID'].tolist()[0]
    kahramanlar.append({
        "episodeTitle": kahraman_adi,
        "seasonNum":id,
        "openingSequenceLocations": muharebeler
    })
dosya_yolu = "jsd3v3.json"

json_data = json.dumps(kahramanlar, indent=4)
print(json_data)
# JSON formatına dönüştürün ve dosyaya yazın
with open(dosya_yolu, 'w', encoding='utf-8') as f:
    json.dump(kahramanlar, f, indent=4)

# JSON formatına dönüştürün ve yazdırın
```
