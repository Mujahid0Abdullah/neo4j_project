```python
import re
def ayir_bolgeler(vefat_yeri):
  """
  Bir vefat yeri bilgisini bölgelere ayırır ve listeye dönüştürür.

  Args:
      vefat_yeri: Vefat yeri bilgisi içeren metin değeri.

  Returns:
      Bölge isimlerini içeren liste.
  """
  bolgeler = []
  # Virgüllerle ayrılmış bölgeleri bulma
  for bolge in vefat_yeri.split(","):
    # Son bölgeyi alma (Karkamış/Gaziantep)
    if re.search(r"/", bolge):
      bolgeler.append(bolge.split("/")[-1])
    else:
      bolgeler.append(bolge.strip())
  return bolgeler

# Yeni bir "Bölge Adları" sütunu oluşturma
df_person["Bölge Adları"] = df_person["Vefat Yeri"].apply(ayir_bolgeler)
```
