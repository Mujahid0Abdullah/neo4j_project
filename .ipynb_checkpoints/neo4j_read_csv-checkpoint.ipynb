```python
from neo4j import GraphDatabase
import pandas as pd
```

```python
URI = "bolt://localhost:7687"
AUTH = ("neo4j", "123123123")
with GraphDatabase.driver(URI, auth=AUTH) as driver:
    driver.verify_connectivity()
```

<h2>PERSONS</h2>

```python
summary = driver.execute_query(
    "match (:Person {adi: $name})",
    name="Abdullah b. Cafer",
    database_="neo4j",
).summary
print("Created {nodes_created} nodes in {time} ms.".format(
    nodes_created=summary.counters.nodes_created,
    time=summary.result_available_after
))
```

```python
records, summary, keys = driver.execute_query(
    "MATCH (p:PERSON) RETURN p.Adi AS name",
    database_="neo4j",
)

for record in records:
    print(record.data()["name"])  # obtain record as dict

# Summary information
print("The query `{query}` returned {records_count} records in {time} ms.".format(
    query=summary.query, records_count=len(records),
    time=summary.result_available_after
))
```

<h2>BATTLES</h2>

```python
Battles= pd.read_csv("battles - battles (3).csv")
Battles
```

```python
for index, row in Battles.iterrows():
    summary = driver.execute_query(
        'CREATE (:BATTLE {Adi: $name, bID: $bID, Sonuc: $Sonuc, Amac: $Amac, Hicri : $Hicri, Miladi: $Miladi , lat:$lat, long: $long, Yer: $Yer,	Savunanlar:$Savunanlar,	Saldıranlar:$Saldiranlar	})',
        name=row["adi"],
        bID= row["bID"],
        Sonuc= row['Sonuç'],
        Amac= row["Amaç"],
        Hicri =int(row["Hicri tarihi"]) ,
        Miladi =row["Miladi tarihi"],
        lat =row["lat"],
        long = row["long"],
        Yer = row["Yer"],
        Savunanlar = row["Savunanlar"],
        Saldiranlar= row["Saldıranlar"],
        database_="neo4j",
    ).summary
    print("Created {nodes_created} nodes in {time} ms.".format(
        nodes_created=summary.counters.nodes_created,
        time=summary.result_available_after))
```
